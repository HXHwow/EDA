```python
def changjvdi(row):
    value = row['常居地']
    if pd.isnull(value):
        return None
    elif value[0] > u'\u4e00' and value[0] <u'\u9fff':
        return value[0:2]
    else:
        return value.split(',')[-1]
```

```python
def aboard(row):
    value = row['常居地']
    if pd.isnull(value):
        return 0
    elif value[0] > u'\u4e00' and value[0] <u'\u9fff':
        return 1
    else:
        return 2
```

```python
douban_df['用户常居地'] = douban_df.apply(changjvdi, axis = 1)
douban_df['常居地类别'] = douban_df.apply(aboard, axis = 1)
```

## 好评，中评，差评中对应的关键字有哪些呢？

```python
import jieba
from wordcloud import WordCloud,ImageColorGenerator
import re
```

```python
# 去除标点符号
def biaodian(text):
    text = re.sub(u'[! ,，。.\n！?？……、;；...：0123456789《》……\、（）=>『]', '',text)
    text = re.sub(u'[A-Za-z=～@\\\[\]\“\”~【／】()]', '',text)
    return text

douban_df['评论'] = douban_df['评论'].apply(biaodian)
```

```python
# 创建停用词表
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords  

stopwords = stopwordslist(r'C:\Users\Hu Xinhua\Desktop\stopwords.txt')

# 增加用户自定义词
jieba.add_word('黄景瑜')
jieba.add_word('战狼')
jieba.add_word('节奏快')
```

```python
# 将评论转化为list格式
wenben = list(douban_df['评论'])
```

```python
# 对评论进行分词
def jiebafenci(text_lst):
    fenci_lst = []
    for line in text_lst:
        line_seged = jieba.cut(line)
        outstr = []
        for word in line_seged:
            if word not in stopwords:
                outstr.append(word)
        fenci_lst.append(outstr)
    text = str()
    for r in fenci_lst:
        for c in r:
            text = text + c + ' ' 
    return text
```

```python
# 输入图片路径绘制云图
def huizhiyuntu(image_path,text):
    color_mask = plt.imread(image_path)
    my_wordcloud = WordCloud(font_path="simsun.ttf",max_words=30,max_font_size=150,background_color="white",
                             mask=color_mask).generate(text)

    image_colors = ImageColorGenerator(color_mask)

    plt.figure(figsize = (12,12))
    plt.imshow(my_wordcloud.recolor(color_func=image_colors))
    plt.axis("off")
    plt.show()
    #my_wordcloud.to_file('保存的图片路径')
```

### 好评关键词

## 成为豆瓣热门评论小窍门探秘

### 什么样的评论更容易成为热评？

```python
from pyecharts import Line

attr = douban_df.index.unique().sort_values()
v1 = douban_df['评论时间'].value_counts().sort_index().values
line = Line('热门评论发布时间')
line.add('',attr,v1,mark_point = ['max'],)
line
```

《红海行动》于2.16号上映。我们可以发现，电影上映前的评论不容易成为电影的热门评论，电影上映当天成为热门评论可能性最高，上映一周内成为热门评论的可能性也比较高，上映20天后也不容易成为热门评论。

如果你想在豆瓣拿电影热评，最好电影上映当天或一周内就去看并评价噢 ！

### 什么时候发布的热评点赞次数会更多？

```python
## 什么时候发布的热评点赞次数会更多？

attr = douban_df.index.unique().sort_values()
v1 = douban_df[['评论时间','有用数']].groupby(['评论时间']).sum().sort_index().values.flatten()

line = Line('热门得到赞时间')
line.add('',attr,v1,mark_point = ['max'])
line
```

电影上映当天给出的赞竟然高达35687次，大大超过其它时间 ~

要想得到的赞多，就电影上映前或当天给出观影评价吧 ~！我们也可以看到，短时间之后，点赞基数就没有那么大了。

## 对热门评论用户做个小研究

### 热门评论的用户加入豆瓣时间分布是什么样的？

```python
# 以年对加入豆瓣时间进行区分
users = douban_df['加入豆瓣时间'].value_counts().sort_index().groupby(pd.TimeGrouper('Y')).sum()

# 将用户加入豆瓣时间转换为年显示
users.index = users.index.strftime('%Y')
```

```python
attr = users.index
v1 = users.values

line = Line('热门评论用户加入豆瓣时间')
line.add('',attr,v1, mark_line = ['average'])
line
```

热门评论中覆盖05年(豆瓣创立年)-18年的用户，其中大部分为05-12年及16年用户。

感觉豆瓣对于用户的持续吸引力还是很足的。   

为什么05-12年用户上热门比较多？
可能参与豆瓣越早的用户其做过的影评较多，对于如何表达自己的感受有更深厚理解。
可能这部分用户豆瓣友圈比较成熟，豆瓣关注者较多，所有受到的赞会更多。
可能这部分用户因为对豆瓣比较关注，所以可能会在影片公映较短时间内就去看电影，并发出评论。
